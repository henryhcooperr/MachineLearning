import dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

data collection

```python
data = pd.read_csv('movies.csv')
data.head()
```

#### Select Relevant Features for Reccomendation

```python
selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
print(selected_features)
```

view the data selected and replace all null values with null strings

```python
for feature in selected_features:
  data[feature] = data[feature].fillna('')
```

```python
combined_features = data['genres']+' '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['director']
```

#### convert text data into feature vectors

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)

print(feature_vectors)
```

Now use cosign similarity to find the similarity scores

```python
similarity = cosine_similarity(feature_vectors)
```

#### Now get a movie name from the user


```python
movie_name = input("Enter your favorite movie name")
```

#### Now create a list of all movie names


```python
movie_names = data['title'].tolist()
print(movie_names)
```

#### Now find the closest match for the movie name given by the user

```python
find_close_match = difflib.get_close_matches(movie_name, movie_names)
close_match = find_close_match[0]
print(close_match)
```

#### Now find a list of similar movies

```python
index_of_the_movie = data[data.title == close_match]['index'].values[0]
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

## sort the movies based on similarity and post the name of the movies based on index

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = data[data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
